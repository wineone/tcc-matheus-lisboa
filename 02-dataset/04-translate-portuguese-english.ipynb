## Notebook to translate the questions of the exams to english, to verify if the models improve

```python
import pandas as pd
import requests
import json
from textblob import TextBlob
import os
```

```python
data = pd.read_csv("data/cleaned/all_questions_df.csv")
data
```

## Translating the text

```python
def concat_line_question(line : pd.Series, text_add_prompt: str):
    """
    """
    text = line['description']
    alternatives = eval(line['alternatives'])
    alternatives = "\n".join(alternatives)
    text = text + "\n" + alternatives
    if text_add_prompt != "":
        text += "\n" + text_add_prompt
    return text

def make_translate(dataframe: pd.DataFrame, output_dir):
    """
    """
    try:
        os.mkdir(output_dir)
    except:
        pass
    for index,line in dataframe.iterrows():
        
        text = concat_line_question(line, "")
        blob = TextBlob(text)
        
        response_content = str(blob.translate(from_lang='pt',to='en'))
        
        name_file = f"{line['year']}-{line['index_question']}.json"
        with open(f"{output_dir}/{name_file}","w") as file:
            file.write(json.dumps({"text": response_content},indent = 2, ensure_ascii = False))
            
        print(f"done index: {index} |",end = "")
    print("Done!")
        
```

```python
make_translate(data, "data/text-blob-translation")
```

## Reading the text translated

```python
def get_response(
    path_jsons: str,
    dataframe: pd.DataFrame
):
    """
    """
    aux = []
    for index,line in dataframe.iterrows():
        name_arq = f"{line['year']}-{line['index_question']}.json"

        with open(f"{path_jsons}/{name_arq}",'r') as file:
            json_response = json.loads(file.read())

        aux.append(json_response['text'])
    dataframe['text_concat_english'] = aux
    
    # concatenating too the text in portuguese
    dataframe[f'text_concat_portuguese'] = dataframe.apply(lambda line: concat_line_question(line,""),axis=1)
            
    return dataframe
```

```python
data_translate = get_response("data/text-blob-translation",data)
data_translate.to_csv("data/cleaned/all_questions_df_english.csv")
```
